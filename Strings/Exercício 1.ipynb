## 1. Cadastro de CPF
Crie um programa para cadastro de CPF de clientes que recebe o CPF em um input box com apenas números.

EX: "insira seu CPF (digite apenas números)"

Caso o usuário digite algo diferente de número ou digite menos de 11 caracteres (tamanho CPF), o programa deve exibir uma mensagem de "Digite seu CPF corretamente e digite apenas números"

```python
cpf_usuario = input("Digite seu CPF (Digite apenas números):")

if len(cpf_usuario) == 11 and cpf_usuario.isnumeric():
    print(f"CPF {cpf_usuario} cadastrado com sucesso!")
else:
    print("Digite seu CPF corretamente e digite apenas números")
```

## 2. Melhorando nosso Cadastro de CPF
Agora, além das validações anteriores, vamos criar um input que permita que o usuários insira pontos, traços e etc...

Nosso programa deve tratar o que o usuário inserir para padronizar apenas o CPF dele em apenas números.

A verificação de tamanho do CPF com 11 caracteres continua válida, mas ela só pode ser feita depois  de retirar todos os pontos, traços e etc... E uma vez retirados, deve sobrar apenas números do CPF. 

No final, nosso programa deve exibir uma mensagem para o usuário, caso ele tenha inserido o CPF inválido ou então deve printar o CPF só com número

```python
cpf_usuario = input("Digite seu CPF: ")
cpf_usuario = cpf_usuario.replace(".", " ") 
cpf_usuario = cpf_usuario.replace("-", " ") 
cpf_usuario = cpf_usuario.strip()

if len(cpf_usuario) == 11 and cpf_usuario.isnumeric():
    print(f"CPF {cpf_usuario} cadastrado com sucesso!")
else:
    print("Digite seu CPF corretamente e digite apenas números")
```

## 3. Cadastro de e-mails
A Hashtag sempre comunica por e-mail. Para isso, verificamos se o e-mail tem @ ou não. Afinal:              
liragmail.com NÃO É e-mail válido               
lira@gmail NÃO É e-mail válido                  
lira@gmail.com é e-mail válido                  

Crie um program que permita o cadastro de nome e e-mail que verifique:                      
Se nome e e-mail foram preenchidos                  
Se o e-mail contém "@" e se depois do @ existe algum "."

```python
nome = str(input("Digite seu nome: "))
email = input("Digite seu e-mail: ")

print(email.find("@"))

if nome and email:
    posicao = email.find("@")
    servidor = email[posicao:]
    if posicao != -1 and "." in servidor:
        print("Cadastro concluído!")
else:
    print("Preencha os dados corretamente!")
```
