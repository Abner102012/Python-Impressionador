# Como descobrir o índice de um item de uma lista?

i = lista.index("item")

Exemplo:                                    
Digamos que você puxou do banco de dados da sua empresa uma lista com todos os produtos que a empresa vende e a quantidade em estoque de todos eles.

```python
produtos = ["tv", "celular", "tablet", "mouse", "teclado", "geladeira", "forno"]
estoque = [100, 150, 100, 120, 70, 90, 80, 15]
```

Nesse caso a lista é pequena para fins didáticos, mas essa lista poderia ter dezenas de milhares de produtos diferentes.                                            
E agora, como eu faço para descobrir a quantidade em estoque do produto geladeira?

```python
produtos = ["tv", "celular", "tablet", "mouse", "teclado", "geladeira", "forno"]
estoque = [100, 150, 100, 120, 70, 90, 80, 15]

i = produtos.index("geladeira")
qtde = estoque[i]

print(i)
print(f"A quantidade de geladeiras no estoque é de {qtde}.")
```

Crie um programa para fazer uma consulta de vendas. O usuário do programa deve inserir o nome do programa e, caso ele não exista na lista, ele deve ser avisado. Caso exista, o programa deve dizer a quantidade de unidades em estoque do produto.

```python
produto = input("Insira o porduto em letra minúscula: ")
if produto in produtos:
    i = produtos.index(produto)
    qtde = estoque[i]
    print(f"Temos {qtde} unidades de {produto} no estoque")
else: 
    print("Produto não encontrado")
```
