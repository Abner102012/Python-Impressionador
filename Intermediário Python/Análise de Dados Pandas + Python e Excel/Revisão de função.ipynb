### 5 Funções essenciais do Pandas que todo Programador PRECISA saber ###

### 1. Describe

```python
import pandas as pd

dados = {
"Nome": ["Alice", "Bob", "Charlie", "David", "Eva"],
"Idade": [25, 30, 35, 40, 45],  
"Salario": [50000, 60000, 70000, None, 90000]
}

tabela = pd.DataFrame(dados)

print(tabela.describe())
```

### 2. Apply

```python
tabela["Salário Ajustado"] = tabela["Salario"].apply(lambda x: x * 1.1)  # Aumenta o salário em 10%
```

### 3. Groupby

```python
medias_area = tabela.groupby("Área")["Salario"].mean(numeric_only=True)
```

### 4. Fillna

```python
tabela["Salario"] = tabela["Salario"].fillna(tabela["Salario"].mean(numeric_only=True))  # Preenche valores nulos com a média
```

### 5. Merge

```python
tabela_areas = pd.DataFrame({
    "Nome": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Área": ["TI", "Marketing", "Financeiro", "RH", "Vendas"],
    "Idade": [25, 30, 35, 40, 45],
    "Salario": [50000, 60000, 70000, None, 90000]
})  
tabela = tabela.merge(tabela_areas, on="Nome", how="left", left_on="Nome", right_on="Nome")  # Faz um merge com base na coluna "Nome"
```
