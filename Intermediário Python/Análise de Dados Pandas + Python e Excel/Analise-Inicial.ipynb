### Puxar base de dados

```python
import pandas as pd

tabela = pd.read_csv('exportacoes_franca.csv')
display(tabela)
```

### Informações Gerais

Quantas linhas têm a tabela? Existe algum valor vazio que deveria ser tratado?

```python
display(tabela.info())
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
tabela_export = tabela[["Year", "US$ FOB"]].groupby('Year').sum()
def formatar(x):
    valor_formatado = f"US$ {x:,.2f}"
    return valor_formatado


tabela_export["US$ FOB"] = tabela_export["US$ FOB"].apply(formatar)
display(tabela_export)
```

### Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
tabela_exportados = tabela[["SH2 Description", "US$ FOB"]].groupby('SH2 Description').sum()
tabela_exportados = tabela_exportados.sort_values(by="US$ FOB", ascending=False)

def formatar(x):
    valor_formatado = f"US$ {x:,.2f}"
    return valor_formatado

tabela_exportados["US$ FOB"] = tabela_exportados["US$ FOB"].apply(formatar)
display(tabela_exportados)
```

### Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
tabela_2020 = tabela.loc[tabela["Year"] == 2020, :]
tabela_cities_2020 = tabela_2020[["City", "US$ FOB"]].groupby('City').sum()
tabela_cities_2020 = tabela_cities_2020.sort_values(by="US$ FOB", ascending=False)
tabela_cities_2020["US$ FOB"] = tabela_cities_2020["US$ FOB"].apply(formatar)
display(tabela_cities_2020)
```

### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
cidade = tabela_cities_2020.index[0]
tabela_cidade = tabela_2020.loc[tabela_2020["City"] == cidade]
tabela_produtos_cidade = tabela_cidade[["SH2 Description", "US$ FOB"]].groupby('SH2 Description').sum()
tabela_produtos_cidade = tabela_produtos_cidade.sort_values(by="US$ FOB", ascending=False)
tabela_produtos_cidade["US$ FOB"] = tabela_produtos_cidade["US$ FOB"].apply(formatar)
display(tabela_produtos_cidade)
```
