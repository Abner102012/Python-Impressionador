# Biblioteca time

### Estrutura: Uma das bibliotecas mais usadas no Python para trabalhar com datas, fazer o programa esperar algum tempo

```python
import time as t
```

- Marco Zero: EPOCH (01 de janeiro de 1970 às 00:00:00)
- time() retorna quantos segundos se passaram desde a EPOCH

```python
print(t.time())
```

Isso já pode ser muito útil para medir o tempo que um código leva para rodar

```python
data_inicial = t.time()
# Código 

# Pegar o tempo depois do código
data_final = t.time()
print(f"Código demorou {data_final - data_inicial:,.4f} segundos")
```

- ctime() retorna a data em string

```python
data_hoje = t.ctime()
print(data_hoje)
```

Fazer o programa esperar alguns segundos (muito útil quando temos que esperar um programa ou uma página carregar)

```python
print("Abrir navegador")
t.sleep(5)
print("Entramos no sistema >:)")
```

Pegar informações de dias, hora, minuto, segundo, tudo detalhado:
- gmtime()
- localtime()

```python
data_gl = t.gmtime()
data_l = t.localtime()

print(data_gl)
print(data_l)
```

Qualquer struct_time permite você pegar as informações da data fazendo:
- gmtime().parametro
- localtime().parametro

```python
dia_today = data_l.tm_mday
mes_hoje = data_l.tm_mon
dia_ano = data_l.tm_yday
print(f"Hoje é dia {dia_today}")
print(f"Esse mês é {mes_hoje}")
print(f"Hoje o dia do ano é {dia_ano}")
```

E para formatar datas para serem exibidas em uma mensagem do jeito que a gente quer?

```python
print(t.strftime("%d %B %Y", data_l))
from locale import *
setlocale(LC_TIME, "pt_BR.UTF-8")
data_l = t.localtime
tempo_formatado = t.strtime("%A, %d de %B de %Y, %H:%M:%S", data_l)
print(tempo_formatado)

# A função strptime() analisa uma string representando um horário de acordo com um formato, devolvendo struct_time
```

E se eu quiser o texto em Português (do Mês e dia da semana). Ai você precisa dizer para o Python o seu "local", usando a biblioteca locale. Link: https://docs.python.org/3/library/locale.html

```python
# A função gmtime() converte um tempo expresso em segundos desde a epoch em um objeto struct_time em UTC
# A gmtime dá o tempo em GMT-0 (da Inglaterra)

# A função mktime() converte um objeto em struct_time em segundos desde a EPOCH. É o inverso da localtime()

tempo_struct = t.localtime()
tempo_seg = t.mktime(tempo_struct)
```
