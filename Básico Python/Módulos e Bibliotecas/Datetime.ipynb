# Guia para a biblioteca `datetime` em Python

A biblioteca `datetime` em Python fornece objetos para manipulação de datas e horas. Aqui está um guia simples para algumas das funções mais úteis deste módulo.



## datetime.datetime.now()

A função `now()` retorna a data e a hora atuais.


```python
from datetime import datetime as date 
import time 
import locale

data_atual = date.now()
print(data_atual)
```

.date() -> pega a data apenas

.time() -> pega o horário apenas

```python
print(data_atual.date())
print(data_atual.time())
```

Para pegar as informações de uma data é mais intuitivo

```python
ano = data_atual.year
mes = data_atual.mounth
dia = data_atual.day
hora = data_atual.minute
segundos = data_atual.second
```

## datetime.date.today()

A função `today()` retorna a data atual.



```python
from datetime import date

dia_hoje = date.today()
print(dia_hoje)
ano = data_atual.year
mes = data_atual.mounth
dia = data_atual.day
hora = data_atual.minute
segundos = data_atual.second

print(f"Data hoje: {dia}/{mes}/{ano}")
```

## Criação de uma data a partir dos valores

Podemos criar um objeto datetime usando o `datetime`. Você pode passar:

- `year`: ano (por exemplo, 2023)
- `month`: mês (1-12)
- `day`: dia (1-31)
- `hour`: hora (0-23)
- `minute`: minuto (0-59)
- `second`: segundo (0-59)
- `microsecond`: microssegundo (0-999999)
- `tzinfo`: fuso horário

```python
dia = 10
mes = 2
ano = 1997

data_analise = date(ano, mes, dia)
print(data_analise)
```

## datetime.timedelta()

A classe `timedelta` é usada para realizar operações com datas (adição e subtração).



```python
from datetime import datetime, timedelta
data_project_inicio = date(2026, 2, 15)

duração = 387
data_final = data_project_inicio + timedelta(days = duração)
print(data_final)
```

## Calcular a diferença entre duas datas

Podemos calcular a diferença entre duas datas subtraindo uma da outra. O resultado será um objeto timedelta.


```python
tempo_project = data_final - data_atual
print(tempo_project.days)
print(type(tempo_project))
```

## Comparação entre datas

Podemos comparar datas usando os operadores de comparação padrão.

Segue uma lógica intuitiva:

    passado < presente < futuro


```python
if data_atual >= data_project_inicio:
    print("Projeto já começou")
else:
    print("Projeto não começou")
```
