## Calculando a idade

Um usuário fornece sua data de nascimento no formato "dd/mm/aaaa". Crie um script Python que calcula a idade do usuário.


```python
from datetime import datetime

data_nascimento_input = input("Informe sua data de nascimento (dd:mm:YYYY): ")
data_nascimento = datetime.strptime(data_nascimento_input, "%d%m%Y")
data_atual = datetime.now()

ano = data_atual.year - data_nascimento.year
mes = data_atual.month
dia = data_atual.day 
mes_user = data_nascimento.month
dia_user = data_nascimento.day

idade = data_atual.year - data_nascimento.year

if mes_user > mes:
    idade -= 1
elif mes_user == mes and dia_user > dia:
    idade -= 1

print(idade)
```
