# Operações com string

str -> transforma número em string                      
in -> verifica se um texto está contido dentro de outro                     
operador + -> concatena string                      
format e {} -> substitui valores                    
%s = substitui textos                   
%d = substitui números decimais                 


```python
faturamento = 1000
custo = 500
lucro = faturamento - custo
print(lucro)
```

#### Uso do str() e do concatenar com +

```python
print("O faturamento da loja foi de " + str(faturamento))
```

#### Uso do format

```python
print(f"O faturamento foi de: R${faturamento}. O custo foi de R${custo}. E o lucro foi de R${lucro}.")
```

#### Uso do %s e %d

```python
print("O faturamento foi de: %d. O custo foi de %d e o lucro foi de %d" % (faturamento, custo, lucro))
```

#### Uso do in 

```python
print("@" in "lira@gmail.com")
```
