##### 12.Faça um programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
salario_hora = float(input("Quanto você ganha por cada hora trabalhada?"))
qtde_horas_mensais = float(input("Quantas horas você trabalha por mês?"))
```

##### Calcule o salário bruto (horas * salário por hora)

```python
salario_hora = float(input("Quanto você ganha por cada hora trabalhada?"))
qtde_horas_mensais = float(input("Quantas horas você trabalha por mês?"))

salario_bruto = salario_hora * qtde_horas_mensais
print(f"O seu salário mensal é R${salario_bruto:.2f}")
```

##### Calcule o desconto do IR(11% do salário bruto)

```python
ir = 11
desconto_ir = salario_bruto * (11 / 100)
print(f"O valor do desconto do IR é de {desconto_ir}")
```

##### Calcule o desconto do INSS(8% do salário bruto)

```python
inss = 8
desconto_inss = salario_bruto * (8 / 100)
print(f"O valor do desconto do INSS é de {desconto_inss}")
```

##### Calcule o desconto do sindicato(5% do salário bruto)

```python
sindicato = 5
desconto_sindicato = salario_bruto * (5 / 100)
print(f"O valor do desconto do sindicato é de {desconto_sindicato}")
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
salario_liquido = salario_bruto/(desconto_ir + desconto_sindicato + desconto_inss)
print(f"O salário líquido é de R${salario_liquido}")
```

##### 13.Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R$80,00. Informe ao usuário a quantidade de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas)

```python
tamanho_area = float(input("Qual é a área em metros quadrados a ser pintada?"))

litro_tinta = tamanho_area / 3

lata = litro_tinta / 18
preco = lata * 80

print(f"A quantidade de latas de tinta são {lata:.2f}. E o preço total é de {preco:.2f}")
```

##### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tamanho_arquivo = float(input("Qual é o tamanho do arquivo em MB?"))
velocidade_internet = float(input("Qual é a velocidade do link de internet em Mbps?"))

tamanho_megabits = tamanho_arquivo / 8

tempo = tamanho_megabits / velocidade_internet

tempo_minutos = tempo / 60

print(f"O tempo esperado é de {tempo}")
```
