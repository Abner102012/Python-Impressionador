# Listas em python          
## Estrutura:   
lista = [valor, valor, valor, valor,...]    

lista é um dos objetos mais importantes de python, por isso vamos trabalhar bastante neles.         
Quando importamos uma base de dados para o python, geralmente ele é lido como uma lista ou variação da lista.           
Listas em python foram feitas para serem homogênias, apesar de aceitarem valores heterogênios.       
Lista[i] -> É o valor de índice i da lista                         
OBS: lembre que no python o índice começa no 0, então o primeiro item de uma lista é o item 0       
Para substituir um valor de uma lista você pode fazer:>
lista[i] = novo_valor    
Exemplos de listas:             

Listas de produtos na loja:

```python
produtos = ["tv", "celular", "mouse", "teclado", "tablet"]
print(produtos[1])
```

Lista de unidades vendidas de cada produto na loja:

```python
vendas = [1000, 1500, 350, 270, 900]
print(vendas[1])
```

#### Nesse caso, as listas funcionam da seguinte forma:


produtos = ["tv", "celular", "mouse", "teclado", "tablet"]
             0        1         2         3          4
vendas = [ 1000,     1500,      350,     270,       900  ]

```python
vendas[3] = 300
print(f"Vendas do priduto {produtos[3]} foram de {vendas[3]} unidades.")
```
