## 1. Criando um mini sistema de controle de estoque
Crie um sistema para ser usado pelo time de controle de estoque de um centro de distribuição.           
Imagine que ao fim de todo o dia, o time conta quantas unidades de produto existem no estoque, se tivermos um estoque abaixo do permitido para aquela categoria, o time deve ser avisado para fazer um novo pedido daquele produto.         
Cada categroria de prodturo tem um estoque mínimo diferente, segundo a regra:           
Alimentos -> Estoque mínimo: 50                     
Bebidas -> Estoque mínimo: 75                   
Limpeza -> Estoque mínimo: 30

Para isso, vamos criar um programa que pede 3 inputs do usuário: Nome do produto, categoria e quantidade atual.

Se o produto tiver abaixo do estoque mínimo, o programa deve printar "solicitar produto à equipe de compras, temos apenas unidades em estoque."

Exemplo: Se o usuário preenche os inputs com : bebidas, dolly, 90. O programa faz nada      
Agora, se o usuário preenche os inputs com: bebidas, guaraná, 60, o programa deve exibir a mensagem: "solicitar guaraná à equipe de compras, temos apenas 60 unidades em estoque."                  

Obs: Lembre de usar o int para transformar o número string em número int                
OBS2: Caso o usuário não preencha alguma das 3 informações, o programa deve exibir uma mensagem para preencher corretamente.

```python
produto = input("Digite qual o produto:")
categoria = input("Digite  qual a categoria do produto:")
Estoque = int(input("Digite a quantidade desse produto no estoque:"))

if produto and categoria and Estoque:
    if categoria == "bebidas":
        if Estoque < 75:
            print(f"Solicitar {produto} à equipe de compras, temos apenas {Estoque} unidades em estoque.")
if produto and categoria and Estoque:
    if categoria == "alimentos":
        if Estoque < 50:
            print(f"Solicitar {produto} à equipe de compras, temos apenas {Estoque} unidades em estoque.")
if produto and categoria and Estoque:
    if categoria == "limpeza":
        if Estoque < 50:
            print(f"Solicitar {produto} à equipe de compras, temos apenas {Estoque} unidades em estoque.")
else: 
    print("Preencha os dados corretamente >:(")
```
