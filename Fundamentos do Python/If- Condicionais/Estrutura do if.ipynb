# Condições no Python -> if
## Estrutura:
### Uso mais simples:

if condição:
   o que fazer caso a condição seja verdadeira

### Exemplo real
Digamos que você trabalha na amazon e está analizando o resultado de vendas dos produtos.

Você precisa criar um programa que vai analisar o resultado de vendas dos produtos da amazon em um mês. Para simplificar, vamos pensar em um único produto: um Iphone.

Meta de vendas do Iphone = 50.000 unidades
Quantidades vendidas no mês = 65.300 unidades

O seu programa deve avisar (usaremso o print por enquanto) caso o produto tenha batido a meta do mês. Então devemos fazer:

Caso o produto tenha batido a meta: "Batemos a meta de vendas do Iphone, vendemos {} unidades"
Se ele não bateu a meta do mês, o seu programa não deve fazer nada

```python
Meta = 50000
qtde_vendida = 65300

if qtde_vendida > Meta:
    print(f"Batemos a meta de vendas do Iphone, vendemos {qtde_vendida} de unidades.")
```

### Tratando a condição falsa:

Quando queremos usar o if, nem sempre queremos apenas analisar o caso verdadeiro, em boa parte das vezes queremos fazer alguma coisa caso contrário a verdadeira e fazer outra coisa se for falsa.

Nesse caso:

if condição:
   o que fazer caso a condição seja verdadeira
else:
   o que fazer caso a condição seja falsa

#### Voltando ao nosso exemplo real da amazon e Iphone, agora o programa deve avisar nos 2 casos.
Caso o produto tenha batido a meta: "Batemos a meta de vendas do Iphone, vendemos {} unidades"
Se ele não bateu a meta do mês: "Infelizmente não batemos a meta, vendemos {} unidades. Sendo que a meta era de {} unidades.


```python
Meta = 50000
qtde_vendida = 15000

if qtde_vendida > Meta:
    print(f"Batemos a meta de vendas do Iphone, vendemos {qtde_vendida} de unidades.")
else:
    print(f"Infelizmente não batemos a meta de Iphones, vendemos {qtde_vendida} unidades. Sendo que a meta era de {Meta} unidades. ;(")
```
