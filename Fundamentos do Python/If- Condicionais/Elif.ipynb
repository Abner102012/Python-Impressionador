## elif
E se temos mais do que um caso de sim ou não?

E se tivermos 3 casos?

Usamos o elif da seguinte forma:

if condição:
    o que fazer se a condição 1 for verdadeira
elif condição2:
    o que fazer se a condição 1 for falsa e a condição 2 for verdadeira
else:
    o que fazer se a condição 1 e a 2 forem falsas

### Exemplo:
Vamos criar um programa para analisar o bônus dos funcionários de uma empresa:

Para os vendedores, a regra é de acordo com as metas:

Se ela vendeu abaixo da meta, não ganha.        
Se ela vendeu acima da meta dela, ganha 3% do valor.        
Se ela vendeu mais do que o dobro da meta dela, ela ganha como bônus 7% do valor.   


Vamos criar um programa que avalia uma pessoa que tinha como meta 20.000 reais e calcular o bônus dela.

```python
meta = 20000
vendas = 25000

if vendas < meta:
    print("Não ganha bônus")
elif vendas > (meta * 2):
    bonus = 0.07 * vendas
    print(f"Ganhou {bonus:.1f} de bônus.")
else:
    bonus = 0.03 * vendas
    print(f"Ganhou {bonus:.1f} de bônus.")
```
